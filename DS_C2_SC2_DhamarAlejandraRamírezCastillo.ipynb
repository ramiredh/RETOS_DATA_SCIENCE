**g**) **Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicaci√≥n entre las columnas daily_vaccionations y daily_vaccionations_raw**

```python
import pandas as pd
vaccin=pd.read_csv("/content/sample_data/country_vaccinations.csv")
vaccin=vaccin.fillna(0)                                              #Necesario primero convertir valores NaN a 0 para que el programa pueda hacer la siguiente operaci√≥n
conDiferencias=vaccin
conDiferencias["diferencias"]=vaccin["daily_vaccinations"]-vaccin["daily_vaccinations_raw"]
conDiferencias.head(8)
```

**i**) **Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada d√≠a (usar la columna vaccines y separar por el car√°cter , )**

```python
import pandas as pd
vacc=pd.read_csv("/content/sample_data/country_vaccinations.csv")
conCantidad=vacc
conCantidad["conVac"]=vacc['vaccines'].str.split(',').str.len()
conCantidad.head(8)
```

**j**) **Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020**

```python
import pandas as pd
vaccination=pd.read_csv("/content/sample_data/country_vaccinations.csv")
antes20=vaccination[vaccination["date"]<'2020-12-20']
antes20.head(8)
```

**k**) **Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer**

```python
import pandas as pd
vaccinati=pd.read_csv("/content/sample_data/country_vaccinations.csv")
vaccinati['pfz']=vaccinati['vaccines'].str.contains('Pfizer', case=False)        #contains

pfizer=vaccinati[vaccinati["pfz"]==True]
pfizer=pfizer.drop(['pfz'],axis=1)
pfizer.head(8)
```

**h**) **Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentaci√≥n de pd.ExcelWriter**

```python
conDiferencias.to_excel("/content/sample_data/resultadosReto.xlsx", sheet_name="conDiferencias")

with pd.ExcelWriter("/content/sample_data/resultadosReto.xlsx", mode="a", engine="openpyxl") as writer:conCantidad.to_excel(writer, sheet_name="ConCantidad")
with pd.ExcelWriter("/content/sample_data/resultadosReto.xlsx", mode="a", engine="openpyxl") as writer:antes20.to_excel(writer, sheet_name="antes20")
with pd.ExcelWriter("/content/sample_data/resultadosReto.xlsx", mode="a", engine="openpyxl") as writer:pfizer.to_excel(writer, sheet_name="pfizer")
```
